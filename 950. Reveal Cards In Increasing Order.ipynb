In a deck of cards, every card has a unique integer.  You can order the deck in any order you want.

Initially, all the cards start face down (unrevealed) in one deck.

Now, you do the following steps repeatedly, until all cards are revealed:

Take the top card of the deck, reveal it, and take it out of the deck.
If there are still cards in the deck, put the next top card of the deck at the bottom of the deck.
If there are still unrevealed cards, go back to step 1.  Otherwise, stop.
Return an ordering of the deck that would reveal the cards in increasing order.

The first entry in the answer is considered to be the top of the deck.

 
```python
Example 1:

Input: [17,13,11,2,3,5,7]
Output: [2,13,3,11,5,17,7]
Explanation: 
We get the deck in the order [17,13,11,2,3,5,7] (this order doesn't matter), and reorder it.
After reordering, the deck starts as [2,13,3,11,5,17,7], where 2 is the top of the deck.
We reveal 2, and move 13 to the bottom.  The deck is now [3,11,5,17,7,13].
We reveal 3, and move 11 to the bottom.  The deck is now [5,17,7,13,11].
We reveal 5, and move 17 to the bottom.  The deck is now [7,13,11,17].
We reveal 7, and move 13 to the bottom.  The deck is now [11,17,13].
We reveal 11, and move 17 to the bottom.  The deck is now [13,17].
We reveal 13, and move 17 to the bottom.  The deck is now [17].
We reveal 17.
Since all the cards revealed are in increasing order, the answer is correct.
                                                 
Note:
1 <= A.length <= 1000
1 <= A[i] <= 10^6
A[i] != A[j] for all i != j
```                                           

https://leetcode.com/problems/reveal-cards-in-increasing-order/

## Best answer so far
* 利用編號為元素產生list
* 把產生的 list 照規則排列
* 利用排列後的 list 對應 deck.sort 產生dict
* `[dict[0], dict[1], dict[2], ..., dict[len(list) - 1]] 即為答案`

```python
from collections import deque

class Solution:
    def deckRevealedIncreasing(self, deck: List[int]) -> List[int]:
        # avoid changing the original deck
        _deck = deck[:]
        # sort the deck
        _deck.sort()
        # make pointer list
        pointer = deque([i for i in range(len(_deck))])
        # move the pointer with the requested method
        moved_pointer = []
        while pointer:
            moved_pointer.append(pointer.popleft())
            if pointer:
                pointer.append(pointer.popleft())
        # after moving, the moved_pointer should relate to _deck
        rel_dict = {}
        for i, j in zip(moved_pointer, _deck):
            rel_dict[i] = j
        # ans should be the val in rel_dict relate to num 0 ~ len(moved_pointer) - 1
        ans = [rel_dict[num] for num in range(len(moved_pointer))]
        
        return ans
```

```python
class Solution:
    def deckRevealedIncreasing(self, deck: List[int]) -> List[int]:
        def sorting(_list, h, t):
            # 開頭指標比結尾指標大，回傳空串列
            if h > t:
                return []
            # 只有一個則回傳
            if h == t:
                return [_list[h]]
            sort_list = []
            left = sorting(_list, h, (h + t) // 2)
            right = sorting(_list, ((h + t) // 2) + 1, t)
            i = 0
            j = 0
            while i < len(left) and j < len(right):
                if left[i] < right[j]:
                    sort_list.append(left[i])
                    i = i + 1
                else:
                    sort_list.append(right[j])
                    j = j + 1
            # 把剩下的走完
            while i < len(left):
                sort_list.append(left[i])
                i = i + 1
            while j < len(right):
                sort_list.append(right[j])
                j = j + 1
                
            return sort_list
                
        h = 0
        t = len(deck) - 1
        sorted_list = sorting(deck, h, t)
        # 逆推，由於最後兩位並沒有在規則裡，所以要例外處理
        rev_list = sorted_list[len(sorted_list)-2:]
        check_list = sorted_list[:len(sorted_list)-2]
        while len(rev_list) < len(sorted_list):
            rev_list = [check_list[-1]] +[rev_list[-1]] + rev_list[:len(rev_list) - 1]
            check_list.pop()
            
        return rev_list
```

```python
class Solution:
    def deckRevealedIncreasing(self, deck: List[int]) -> List[int]:
        if len(deck) < 2:
            return deck
        # 避免修改到原本list
        check_list = deck[:]
        check_list.sort()
        # 逆推
        # 由於最後兩位數並沒有在規則裡，要額外處理
        rev_list = []
        for i in range(2):
            rev_list = [check_list.pop()] + rev_list

        # check_list內還有東西就要做
        while len(check_list) > 0:
            # pop 出來再加回去可以不用slice
            temp = [rev_list.pop()]
            rev_list = [check_list.pop()] + temp + rev_list

        return rev_list
    ```
