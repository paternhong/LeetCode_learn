Given the root node of a binary search tree (BST) and a value to be inserted into the tree, insert the value into the BST. Return the root node of the BST after the insertion. It is guaranteed that the new value does not exist in the original BST.  
Note that there may exist multiple valid ways for the insertion, as long as the tree remains a BST after insertion. You can return any of them.

For example, 
```python
Given the tree:
        4
       / \
      2   7
     / \
    1   3
And the value to insert: 5
You can return this binary search tree:

         4
       /   \
      2     7
     / \   /
    1   3 5
This tree is also valid:

         5
       /   \
      2     7
     / \   
    1   3
         \
          4
```

## Answer 1
```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None
class Solution:
    def insertIntoBST(self, root: TreeNode, val: int) -> TreeNode:
        if not root:
            # 當下沒有root，以val為值生成Node回傳
            return TreeNode(val)
        if val >= root.val:
            # 比當下大或等於，則新增到right child
            root.right = self.insertIntoBST(root.right, val)
        else:
            # 比當小，則新增到left child
            root.left = self.insertIntoBST(root.left, val)
        
        return root
```

## Answer 2
```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def insertIntoBST(self, root: TreeNode, val: int) -> TreeNode:
        cur_node = root
        while cur_node:
            # 當下的node會變成前一個node，用來定位'左右'
            pre_node = cur_node
            # 大於等於當下 node.val, 往右邊跑
            if val >= cur_node.val:
                cur_node = cur_node.right
            # 小於當下 node.val, 往左邊跑
            else:
                cur_node = cur_node.left
        if not pre_node.right and val >= pre_node.val:
            pre_node.right = TreeNode(val)
        else:
            pre_node.left = TreeNode(val)
```

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def insertIntoBST(self, root: TreeNode, val: int) -> TreeNode:
        if not root:
            return root
        curNode = root
        while True:
            if val < curNode.val:
                if curNode.left:
                    curNode = curNode.left
                else:
                    curNode.left = TreeNode(val)
                    break
            else:
                if curNode.right:
                    curNode = curNode.right
                else:
                    curNode.right = TreeNode(val)
                    break
        return root
```
